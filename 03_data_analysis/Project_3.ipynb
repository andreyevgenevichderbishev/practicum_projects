### Предобработка данных

#### Типы данных и названия столбцов ####

- Преобразуем название столбца к обшепринятому формату `cityCenters_nearest` к `city_center_nearest`.

- Можно заметить что столбцы которые по логике содержат целые числа (`balcony`,`city_center_nearest`,`airports_nearest`,`parks_around3000`,`parks_nearest`,`ponds_around3000`,`ponds_nearest`,`days_exposition`) содержат действительные числа. Но смысла менять тип нет.

#### Площадь кухни и жилая площадь.

Предположим что жилая площадь это доля полной площади, а площадь кухни это доля нежилой площади. Посмотрим гистограммы.

```python
data['living_area'].hist(bins=100)
```

```python
data['kitchen_area'].hist(bins=50)
```

Распределение жилой площади похоже на распределение полной площади. А вот кухня в 100 квадратов даже в 200м квартире это странно. Нарисуем диаграмму размаха для площади кухни.

```python
data['kitchen_area'].plot(kind='box',ylim=(0,100))
```

```python
data.loc[data['kitchen_area']>35,'kitchen_area'].count()/data['kitchen_area'].count()
```

Опять выберем интервал до 35м.кв. с запасом т.к. выбросов много и удалим аномальные значения.

```python
data = data.loc[(data['kitchen_area']<=35)|(data['kitchen_area'].isna())]
```

Строим гистограмму без аномалий.

```python
data['kitchen_area'].hist(bins=30)
```

Проведем так же простую проверку коректности данных по площадям. Кухня это часть нежилой площади т.е. полная площадь меньше жилой + кухня. Если это не так данные не коректны и их нужно удалить.

```python
data.loc[(data['kitchen_area']+data['living_area'])>data['total_area'],'kitchen_area']\
.count()/data['kitchen_area'].count()
```

Да такие записи есть и их 0.3% . Удаляем их.

```python
data = data.loc[(data['kitchen_area'].isna())|\
         (data['living_area'].isna())|\
         ((data['kitchen_area']+data['living_area'])<=data['total_area'])]
```

В графе площадь кухни присутствуют пропущенные значения. Их можно было бы заменить медианными значениями, но дальше мы хотим искать зависимости других величин он площади кухни. Заменив на медиану мы испортим распределение и можем изменить зависимости. По этому заполнять пропуски не будем.

##### Этажи

Пропуски есть в графе `floors_total`. этаж же присутствует во всех записях.

```python
data['floors_total'].isna().sum()
```

Пропусков мало их можно выкинуть без вреда для статистики.

Остальные данные проверим на коректность. Этаж должен быть не больше висла этажей в доме. А максимальный этаж не больше 37 (самое высокий жилой этаж в СПБ и окрестностях.)

В ходе дальнейшего анализа было определено что цена объекта в большой степени зависит от его полной площади и связанными с ней величинами т.е. жилой площадью , площадью кухни , количеством комнат. И растет линейно при росте последних. Другим важным параметром служит этаж объекта. Первые и последние этажи в среднем несколько дешевле остальных, причем первые дешевле последних. От дня недели и месяца цена не зависит, а вот от года зависимость есть, но она определяется внешними макроэкономическими факторами.
Так же была выявлена зависимость средней цены квадратного метра от расположения объекта.
Самый дорогой в среднем квадратный метр в СПБ, цены в пригородах значительно ниже , так например, в Выборге средняя цена м.кв. почти в 2 раз ниже чем в СПБ.
В самом же СПБ средняя цена кв.м. зависит от расстояния до центра города. Стоимость кв.м. в центре почти на треть больше ,чем на окраинах города. При этом в центре есть значительное количество очень дорогих кв.м. близкие к значением которые считаем аномальными в среднем для региона.
Таким образом были выведены основные факторы по которым можно пытаться предсказывать стоимость квартиры.
