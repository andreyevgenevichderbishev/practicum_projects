#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

# План работы

Работа устроена следующим образом:
- В первой главе загружаем данные 
- Во второй главе проводим предобработку, ресемплируем временной ряд и создаем дополнительные признаки
- В 3 главе проводим анализ данных.Смотрим корреляции и диаграммы рассеянья.
- Глава 4 обучение моделей.
- 5 глава - тестируем модель и сравниваем результаты с простейшими вариантами предсказаний. Медианой и предыдущим значением.
- В последней главе делаем общие выводы по результатам работы.

### Новые признаки

Создадим дополнительные признаки отстающие значения и скользяшее среднее, а так же , день  и день недели. Предварительно ресемплируем по часу.

```python
df.sort_index(inplace=True)
df = df.resample('1H').sum()
print(df.head(1).index)
print(df.tail(1).index)
```

Видно, что все данные за пол года. 

```python
def make_features(data, max_lag, rolling_mean_size):
    data['weekday'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    data['rolling_std'] = data['num_orders'].shift().rolling(rolling_mean_size).std()
    
    data = data.drop(data.head(max(max_lag, rolling_mean_size)).index,inplace=True)
    
    return data
```

```python
make_features(df,MAX_LAG,ROLLING_MEAN_SIZE)
df.head()
```

Построим так же графики автокорреляции и частичной автокорреляции

```python
def autocor(lags):
    plot_pacf(df['num_orders'], alpha=None,lags=lags,method='ywm').set_size_inches(12, 4)
    plt.title('Частичная автокорреляция')
    plt.xlabel('Время ч.')
    plt.ylabel('Коэфициент автокорреляции')
    plt.show()

    plot_acf(df['num_orders'],lags=lags, alpha=None).set_size_inches(12, 4)
    plt.xlabel('Время ч.')
    plt.ylabel('Коэфициент автокорреляции')
    plt.title('Автокорреляция')
    plt.show()
autocor(30)
```

## Обучение

Зазбиваем данные на тестовую и обучающие выборки.

```python
num = df.drop(['num_orders','rolling_std','hour','weekday'],axis=1).columns.tolist()
cat = ['hour','weekday']

X = df.drop(['num_orders','rolling_std'],axis=1)
y = df['num_orders']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = TEST_SIZE,
    shuffle=False, 
    random_state = RANDOM_STATE)

X_train.shape, y_train.shape
```

Создадим пайплайн для выбора лучшей модели и ее гиперпараметров.

```python
# создаём пайплайн для подготовки признаков из списка cat: OHE-кодирование
ohe_pipe = Pipeline(
    [
        (
            'ohe', 
            OneHotEncoder(drop='first',handle_unknown='ignore', sparse_output=True)
        )
    ]
)

# создаём пайплайн для подготовки данных gb
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, cat),
        ('num', RobustScaler(), num)
    ], 
    remainder='passthrough'
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final= Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LGBMRegressor(random_state=RANDOM_STATE))
    ]
)
```

```python
param_grid = [
    
    # словарь для модели KNeighborsRegressor()
    {
        'models': [KNeighborsRegressor()],
        'preprocessor__num': ['passthrough'],
        'models__n_neighbors': range(20,25)
    },
    
    # словарь для модели LinearRegression()
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler()] 
    },
    
    # словарь для модели LGBMRegressor()
    {
        'models': [LGBMRegressor(random_state=RANDOM_STATE)],
        'models__num_leaves': range(5,9),
        'models__max_depth': range(5,9), 
        'models__learning_rate': [0.1,0.15],
        'preprocessor__num': ['passthrough'] 
    }
]
```

```python
%%time
tscv = TimeSeriesSplit(n_splits=5)
randomized_search = RandomizedSearchCV(
        pipe_final, 
        param_grid, 
        cv=tscv,
        n_iter=20,
        scoring='neg_root_mean_squared_error',
        random_state=RANDOM_STATE,
        n_jobs=1
    )


randomized_search.fit(X_train, y_train)
print ('Метрика лучшей модели при кросс-валидации:', -randomized_search.best_score_)
randomized_search.best_estimator_
```

Рейтинг моделей.

```python
result = pd.DataFrame(randomized_search.cv_results_)
result[
    ['rank_test_score', 'param_models', 'mean_test_score','params']
].sort_values('rank_test_score')
```

## Тестирование

Для начала посмотрим какое качество будет иметь простейшее предсказание
- медианой за прошлые значения.
- значение за предыдущий час.

```python
y_pred_previous = y_test.shift()
y_pred_previous.iloc[0]=y_train.iloc[len(y_train)-1]

print("Для предыдущего значения RMSE:", (mean_squared_error(y_pred_previous,y_test))**0.5)
```

```python
y_pred_median = [y_train.median()]*len(y_test)
print("Для медианного значения RMSE:", (mean_squared_error(y_pred_median,y_test))**0.5)
```

Посмотрим на метрику RMSE обученной модели на тестовой выборке.

```python
y_test_pred = randomized_search.best_estimator_.predict(X_test)
print(f'Метрика RMSE на тестовой выборке линейная регрессия: {(mean_squared_error(y_test, y_test_pred))**0.5}')
```

Посмотрим распределение остатков.

```python
residual = (y_test-y_test_pred)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].hist(residual,bins=30)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_ylabel('Число точек')
axes[0].set_xlabel('Остатки')

axes[1].scatter(y=residual,x=y_test_pred, alpha=0.5)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
print(residual.describe())
```

Остатки не сильно равномерные и есть несколько больших выбросов.

Посмотрим так же сравнение предсказания и фактических данных.

```python
dif = y_test.reset_index()
dif['pred'] = y_test_pred
dif = dif.set_index('datetime')
dif.plot(figsize=(10,10))
plt.legend(['Фактические', 'Предсказанные'])
plt.title('Фактические и предсказанные значения в тестовой выборке')
plt.xlabel('Время')
plt.ylabel('Число заказов')
plt.show()
```
