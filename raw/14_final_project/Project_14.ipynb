# Модель предсказания оттока клиентов.

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Ваша задача — обучить на этих данных модель для прогноза оттока клиентов.

Описание услуг

Оператор предоставляет два основных типа услуг: 
- Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:
- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

### Зависимости категорийных признаков.

````python
cat = df.select_dtypes(exclude=['float','bool','int']).columns.tolist()
cat = cat+['Extra']

for i in cat :
    
    (df.pivot_table(index=i,values='Target',aggfunc='count')\
        .sort_values(by='Target',ascending=False)/df.shape[0])\
        .plot(kind='bar',rot='45',legend=False)
    plt.xlabel(i)
    plt.title(f'Зависимость доли пользователей от {i}')
    plt.ylabel('Доля пользователей')
    
    temp = df.copy()
    temp['NTarget'] = temp['Target']==False
    temp2 = temp.pivot_table(index=i, values=['Target','NTarget'], aggfunc=['sum'])
    temp2['sum2'] = temp2[('sum', 'NTarget')]+temp2[('sum', 'Target')]
    temp2[('sum', 'NTarget')]=temp2[('sum', 'NTarget')]/temp2['sum2']
    temp2[('sum', 'Target')]=temp2[('sum', 'Target')]/temp2['sum2']
    temp2[[('sum', 'NTarget'),('sum', 'Target')]].plot(kind='bar',stacked=True)
    plt.title(f'Доля ушедших пользователей в разрезе по {i}')
    plt.xlabel(i)
    plt.legend(['Остались','Ушли'],title = 'Статус')
    plt.ylabel('Доля пользователей')
    
    plt.show()
````

## Подготовка данных

Разобьем выборки на обучающую и тестовую в отношении 3:1 с учетом стратификации. Как видели выше у целевого признака есть дисбаланс классов. Но для оценки моделей будет использоваться метрика ROC-AUC которая не чувствительна к дисбалансу. 

````python
# Создадим явно списки численных и категорийных величин. 
num = ['MonthlyCharges','TotalCharges','Days','Extra']
cat = ['Type','PaperlessBilling','PaymentMethod','SeniorCitizen','Partner',
       'InternetService','MultipleLines']

for i in cat:
    df[i] = df[i].astype('category')


X = df[cat+num]
y = df['Target']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = TEST_SIZE,
    stratify = df['Target'],
    random_state = RANDOM_STATE)

X_train.shape, y_train.shape
````

Для кодирования категорийных признаков будем использовать ohe-енкодер для Логистической регрессии и OrdinalEncoder для случайного леса и градиентного бустинга.

````python
# создаём пайплайн для подготовки признаков из списка cat: OHE-кодирование 
# для логистической регресии
ohe_pipe = Pipeline(
    [
        (
            'ohe', 
            OneHotEncoder(drop='first',handle_unknown='error', sparse_output=True)
        )
    ]
)


# и OrdinalEncoding для леса и градиентного бустинга

ord_pipe = Pipeline(
    [
        ('ord',  OrdinalEncoder(handle_unknown='error'
                )
        )
    ]
)
````

## Обучение модели

### Логистическая регрессия

Создаем пайплайн для логистической регрессии и с помощью RandomizedSearchCV ищем оптимальные гиперпараметры.

````python
# создаём пайплайн для подготовки данных 
# LinearRegression()
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, cat),
        ('num', StandardScaler(), num)
    ], 
    remainder='passthrough'
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final= Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear',class_weight='balanced'
        ))
    ]
)
````

</div>

Думаю там не столь серьезные выбросы. Не 1.5 межквартильного размаха , а 2 и довольно много точек. Т.е. не сколько выбросы сколько плавный хвост распределения. И только в одном численном признаке из 3. По этому StandardScaler() тут справится. (Практически от замены одного на другой метрика ни как не меняется.)

````python
 # словарь для модели LogisticRegression()
param_grid = [    
    {
        'models__penalty':['l1','l2'],
        'models__C':range(1,10)
    }
]
````

````python
randomized_search = RandomizedSearchCV(
        pipe_final, 
        param_grid, 
        cv=5,
        n_iter=18,
        scoring='roc_auc',
        random_state=RANDOM_STATE,
        n_jobs=-1
    )


randomized_search.fit(X_train, y_train)
print ('Метрика лучшей модели при кросс-валидации:', randomized_search.best_score_)
display(randomized_search.best_estimator_)
````
